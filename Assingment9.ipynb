Q6. List down some best practices for exception handling in python.

```python
#CODE FOR Q6.
# program for custom expection handling in python 
import logging
# these lines of code used for create a log file to track all the activity for this program
logging.basicConfig(filename="cust1.log", level=logging.INFO,format="%(asctime)s,%(message)s") 
class detact(Exception):
    def __init__(self,msg):
        self.msg = msg
def dettact(age):
    if age < 0 :
        logging.info("Here is error is record age is negative")
        raise detact("Age is Negative")
    elif age > 200:
        logging.info("Here is error is record age is Too High")
        raise detact("Age is too high")
try:
    dettact(240)
except detact as e:
    print(e)
logging.shutdown()
```
